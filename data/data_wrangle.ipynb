```python
def api_as_dict(result):
  api_dict = {}
  lat = None
  long = None
  full_address = None
  address_3 = None
  address_2 = None
  address_1 = None 
  country = None

  if len(result) != 0:
    result = result[0]

    for d in result:
      if d == 'address_components':
        for q in result['address_components']:
          if q['types'][0] == 'administrative_area_level_3':
            address_3 = q['long_name']
          elif q['types'][0] == 'administrative_area_level_2':
            address_2 = q['long_name']
          elif q['types'][0] == 'administrative_area_level_1':
            address_1 = q['long_name']
          elif q['types'][0] == 'country':
            country = q['long_name']
      if d == 'geometry':
        lat = result['geometry']['location']['lat']
        long = result['geometry']['location']['lng']
      if d == 'formatted_address':
        full_address = result['formatted_address']

  api_dict['lat'] = lat
  api_dict['long'] = long
  api_dict['full_address'] = full_address
  api_dict['address_3'] = address_3
  api_dict['address_2'] = address_2
  api_dict['address_1'] = address_1
  api_dict['country'] = country

  return api_dict
  
```
